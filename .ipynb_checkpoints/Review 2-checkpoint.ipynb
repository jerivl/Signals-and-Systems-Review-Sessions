# Review 2 - Fourier Analysis
The frequency domain often gives a more interpretable description of a signal than its time-domain representation.

## Periodic Signals
Many of the signals we look at are periodic, i.e. $ f(t) = f(t + T)$ for some real number $T$

As proved in mathematical analysis: any periodic signal can be represented as an infinite sum of sinusoids (or complex exponentials).

$$\begin{align*}
   f(t) & = \sum_{n=-\infty}^{\infty}c_ne^{jn\omega_0t }
\end{align*}$$

Notice that this sum converges to any general, aperiodic $f(t)$ only when we have an **infinite** number of terms. 
However, for many purposes a finite sum is a good enough approximation.

## Fourier Series
So we have several conventions for writing this finite sum for some periodic function $f(t)$
$$\begin{align*}
   f(t) & = \sum_{n=-N}^{N}c_ne^{jn\omega_0t } & (1)\\ \\ & = {a_0} + 2\sum_{n=1}^{N}B_ncos(n\omega_0t)+C_nsin(n\omega_0t)] & (2)
\end{align*}$$

Equation (1) is sometimes called the exponential fourier series (EFS)

Equation (2) is sometimes called the complex fourier series (CFS)

If we think of $f(t)$ as a sound wave, the fourier series equations say that our signal is made of different pitches or tones.

## Common Fourier Series
To see this: consider the Fourier transform of a sine function 
$$\begin{align*}f(t)= sin(\omega_0t)=cos(\omega_0t-\frac{\pi}{2})& \quad \leftrightarrow & F(j\omega) &= \frac{\pi}{j}[\delta(\omega - \omega_0) - \delta(\omega + \omega_0)] \\ \quad & \quad & &= \pi{e^{-j\frac{\pi}{2}}}\delta(\omega - \omega_0) + \pi{e^{j\frac{\pi}{2}}}\delta(\omega + \omega_0)
\end{align*}$$

```python
# TODO: Add sliders
def cos_demo(omega, amplitude, phase):

    fig, axs = plt.subplots(2, 2)
    gs = axs[0, 0].get_gridspec()
    # remove the underlying axes
    for ax in axs[0:, 0]:
        ax.remove()
    axbig = fig.add_subplot(gs[0:, 0])
    t = np.linspace(0,1,10000)
    axbig.plot(t,amplitude*np.cos(omega*t+phase))
    axbig.set_xlim([0,1/10])
    axbig.set_ylim([-2, 2])
    axbig.set_title('Time-Domain Signal')
    axbig.set_xlabel('Time (t)')

    f = np.linspace(-3000*2*np.pi,3000*2*np.pi,100000)
    M = np.zeros(100000)
    P = np.zeros(100000)

    M[[np.argmin(abs(f-2*np.pi*omega)),np.argmin(abs(f+2*np.pi*omega))]] = amplitude*np.pi
    P[np.argmin(abs(f+2*np.pi*omega))] = phase
    P[np.argmin(abs(f-2*np.pi*omega))] = -phase

    _ = axs[0,1].stem(f, M, use_line_collection = True)
    _ = axs[0,1].set_title('Magnitude Spectrum')
    _ = axs[0,1].set_yticks(np.arange(0,2*np.pi+2*np.pi/2,np.pi/2))
    _ = axs[0,1].set_yticklabels(['0','Ï€/2','Ï€','3Ï€/2','2Ï€','...'])
    _ = axs[0,1].set_xticks([])
    _ = axs[1,1].stem(f, P, use_line_collection = True)
    _ = axs[1,1].set_ylim([-np.pi,np.pi])
    _ = axs[1,1].set_title('Phase Spectrum')
    _ = axs[1,1].set_yticks(np.arange(-np.pi,np.pi+np.pi/4,np.pi/4))
    _ = axs[1,1].set_yticklabels(['-Ï€','-3Ï€/4','-Ï€/2','-Ï€/4','0','Ï€/4','Ï€/2','3Ï€/4','Ï€'])
    _ = axs[1,1].set_xticks([f[np.argmin(abs(f+2*np.pi*omega))],0,f[np.argmin(abs(f-2*np.pi*omega))]])
    _ = axs[1,1].set_xticklabels([-omega,0,omega]) # TODO: fix axis
    _ = axs[1,1].set_xlabel('Frequency (Ï‰)')

def play_tone(omega, duration = 1.0, sample_rate = 22050):
    t = np.linspace(0, duration, int(duration*sample_rate), endpoint=False) # time variable
    x = 0.5*np.cos(omega*t) # pure sine wave at 440 Hz
    ipd.Audio(x, rate=sample_rate) # create sound from numpy array
```

```python
from sonopy import mel_spec
import numpy as np
import seaborn as sns
import pandas as pd

sample_rate = 16000
t_frame = 0.020
t_stride = 0.010

lengths = 48159+1
y = np.zeros(lengths)
for i in range(lengths):
    audio = np.zeros(i)
    x = mel_spec(audio, sample_rate, window_stride=(int(sample_rate*t_frame), int(sample_rate*t_stride)), fft_size=512, num_filt=26).astype('float32')
    y[i] = x.shape[0]
    
df = pd.DataFrame(dict(length=np.arange(lengths),
                       features=y))
g = sns.relplot(x="length", y="features", kind="line", data=df)
idx = np.argwhere(np.array([y[i] - y[i-1] for i in range(1,len(y))]) == 1)
```

```python
mean = 2.164693
var = 7.879409
scale = var/mean
shape = mean/scale

import time
t0 = time.time()
times = []
max_features_len = 30000
min_audio_len = t_frame*sample_rate-1
max_audio_len = t_frame*sample_rate-1 + max_features_len*t_stride*sample_rate
quant = np.arange(min_audio_len, max_audio_len, t_stride*sample_rate).astype(np.int)
for _ in range(10000):
    length = np.random.gamma(shape,scale)
    audio = np.zeros(int(length*sample_rate))
    x = mel_spec(audio, sample_rate, window_stride=(int(sample_rate*t_frame), int(sample_rate*t_stride)), fft_size=512, num_filt=26).astype('float32')
    y = x.shape[0]
    
    t1 = time.time()
    
    num_features = np.searchsorted(quant, len(audio), side='left')
    
    trimmed = quant[num_features]
    audio = np.zeros(trimmed)
    x = mel_spec(audio, sample_rate, window_stride=(int(sample_rate*t_frame), int(sample_rate*t_stride)), fft_size=512, num_filt=26).astype('float32')
    y_hat = x.shape[0]
    
    assert y_hat == y
    assert len(audio) in quant
    times.append(time.time()-t1)
print(time.time()-t0)

sum(times)
```

```python
mean = 2.164693
var = 7.879409
scale = var/mean
shape = mean/scale

import time
t0 = time.time()
times = []
max_features_len = 30000
min_audio_len = t_frame*sample_rate-1
max_audio_len = t_frame*sample_rate-1 + max_features_len*t_stride*sample_rate
quant = np.arange(min_audio_len, max_audio_len, t_stride*sample_rate).astype(np.int)
for _ in range(100):
    length1 = np.random.gamma(shape,scale)
    length2 = np.random.gamma(shape,scale)
    audio1 = np.random.randint(-32768, high=32767, size=int(length1*sample_rate))
    audio2 = np.random.randint(-32768, high=32767, size=int(length2*sample_rate))
    x1 = mel_spec(audio1, sample_rate, window_stride=(int(sample_rate*t_frame), int(sample_rate*t_stride)), fft_size=512, num_filt=26).astype('float32')
    x2 = mel_spec(audio2, sample_rate, window_stride=(int(sample_rate*t_frame), int(sample_rate*t_stride)), fft_size=512, num_filt=26).astype('float32')
    
    y1= x1.shape[0]
    y2= x2.shape[0]
    
    t1 = time.time()
    
    num_features1 = np.searchsorted(quant, len(audio1), side='left')
    num_features2 = np.searchsorted(quant, len(audio2), side='left')
    
    
    trimmed1 = quant[num_features1-1]+1
    trimmed2 = quant[num_features2-1]+1
    audio1 = audio1[:trimmed1]
    audio2 = audio2[:trimmed2]
    audio = np.concatenate((audio1,audio2))
    
    x1 = mel_spec(audio1, sample_rate, window_stride=(int(sample_rate*t_frame), int(sample_rate*t_stride)), fft_size=512, num_filt=26).astype('float32')
    x2 = mel_spec(audio2, sample_rate, window_stride=(int(sample_rate*t_frame), int(sample_rate*t_stride)), fft_size=512, num_filt=26).astype('float32')
    
    y1_hat = x1.shape[0]
    y2_hat = x2.shape[0]
    
    assert y1 == y1_hat
    assert y2 == y2_hat
    
    x = mel_spec(audio, sample_rate, window_stride=(int(sample_rate*t_frame), int(sample_rate*t_stride)), fft_size=512, num_filt=26).astype('float32')
    y_hat = x.shape[0]
    
    assert y_hat == (y1_hat + y2_hat)
    
    times.append(time.time()-t1)
print(time.time()-t0)

sum(times)
```
